Secondly, each building's site might be understood as a reflection of the broader, geographically distributed seismic activity. For example, buildings constructed near fault lines might offer insights into the local seismic behavior, tectonic movements, or even soil liquefaction patterns. ### 3.1. Hexbin : 

Our problem is trying to find a density of the spatial distribution. One way to solve this is using hexbin function by python, dividing are a to hexagons. 
Hexbins are similar to grids and grids are as similar to bar charts where each bar counts things in a particular space. Just like we choose how many bars or categories we want in a bar chart, we decide the number of "buckets" or areas in a grid. This way of counting is quick and can be very detailed. But there's a downside: it can simplify and change how we view things that naturally spread out without clear boundaries.

In our example, in order to demonstrate better grid size is chosen as 150 so that different city centers also can appear on the moap.

Below graph reflect us the density of the all buildings on the map.Our data is not homogenous as builgings are in the city centers.

```python
# Set up figure and axis
f, ax = plt.subplots(1, figsize=(20, 20))
# Generate and add hexbin with 150 hexagons in each
# dimension, no borderlines, half transparency,
# and the reverse viridis colormap
hb = ax.hexbin(
    data["latitude"],
    data["longitude"],
    gridsize=150,
    linewidths=0,
    alpha=0.5,
    cmap="viridis_r",
)
# Add basemap
ctx.add_basemap(
    ax, source=ctx.providers.CartoDB.Positron
)
# Add colorbar
plt.colorbar(hb)
```

### 3.2 KDE(Kernel Density Estimation):

Another way to show density is using kernel density histogram. It's like a smooth, even blanket laid over a space. Instead of just counting things in square or hexagon areas, KDE uses special functions, like mini magnets, that pull points towards them. The closer a point is to the magnet, the stronger the pull. All these pulls together make a smooth, continuous surface that shows the likelihood of finding something in any spot. The most popular type of this "magnet" or function is called Gaussian, which pulls points in a bell-curve shape. So, with KDE, we get a smooth map showing where things are likely found.

```python
# Set up figure and axis
f, ax = plt.subplots(1, figsize=(12,9))
# Generate and add KDE with a shading of 150 gradients
# coloured contours, 75% of transparency,
# and the reverse viridis colormap
sns.kdeplot(
    x="latitude",
    y="longitude",
    data=data,
    n_levels=100,
    fill = False,
    alpha=0.55,
    cmap="viridis_r",
)
# Add basemap
ctx.add_basemap(
    ax, source=ctx.providers.OpenStreetMap.Mapnik
)
```

## 3.3. Centrography : 

Centrography is a study focused on analyzing the centrality of a point pattern. It evaluates the overall location and spread of a pattern, similar to how the mean is used in measures of central tendency. This method helps summarize spatial distributions into more concise information sets, like a single point. Centrography uses various indices to determine the location of a point pattern, its clustering intensity around the center, and the irregularity of its shape.


## 3.4 Tendency:
Tendency is a way to find the middle of point patterns, finding the "heart" of the data. The center of mass is like the average spot of marked data points. If the data points don't have marks, this "heart" or center is just the average location. There's also a "middle" point where half the data is around it, called the median center. We can use centography function of the pointpats library.


Below we are going to examine Centrography and tendency together. We are trying to understand reference to 3 important seismic activity in the disaster day. 

```python

#calculating the mean center of the data destroyed and damaged data
mean_center_destroyeddamaged = centrography.mean_center(filtered_data[["latitude", "longitude"]])
med_center_destroyeddamaged = centrography.euclidean_median(filtered_data[["latitude", "longitude"]])
#calculating the mean center of all the data 
mean_center_data= centrography.mean_center(data[["latitude", "longitude"]])
med_center_data = centrography.euclidean_median(data[["latitude", "longitude"]])


# Locations of the important earthquakes over magniture 6 on the disaster day.
Pazarcik_point = 37.288, 37.043	
Nurdagi_point = 37.304, 36.920
elbistan_point = 38.089, 37.239
```

### 3.5. Randomness and Clustering:

Spatial statistics often look at the evenness of point distributions: whether points group closely together or spread out across an area. This deals with the intensity or spread of the point pattern. The aim is to understand the overall clustering in the pattern, and much work has been done on this topic. Two main methods to study clustering are:

1.Quadrat statistics: This method divides data into small areas called quadrants. These areas are then studied to see how evenly points are distributed across them.


2.Techniques based on Ripley (1988): These focus on measuring the distances between points in a pattern.

```python
#filtering the data to include only locations
filtered= pd.DataFrame(filtered_data[['longitude', 'latitude']])
#creating pp object to understand point pattern characteristics
pp = PointPattern(filtered.values)

qstat = QStatistic(pp)
print(f'gstat = {qstat.chi2_pvalue}')
QStatistic.
qstat.plot()
```

The QStatistic computes the Q-statistic for a point pattern, which is used to evaluate the randomness of the spatial distribution of points. If the points are randomly distributed, the Q-statistic will be close to 1. If the points tend to be clustered, the Q-statistic will be greater than 1, and if they tend to be uniformly distributed, the Q-statistic will be less than 1.

If the chi2_pvalue is less than 0.05,(in this case) it indicates that the spatial distribution of the points in the filtered dataset is statistically significantly different from a random distribution.

The plot generated by qstat.plot() will provide a visual representation of this distribution, allowing you to determine if the points are more clustered or more evenly spread out than what would be expected under a random distribution.

The actual interpretation depends on the value of chi2_pvalue and the visual characteristics of the plot. For instance, if the p-value is very low and the Q-statistic value is significantly greater than 1, it suggests a clustered pattern.


To compare with the real distribution, we can create a random distribution of the points to see what os 

```python

# Compute bounding box from the data using DataFrame methods
min_x = filtered['longitude'].min()
max_x = filtered['longitude'].max()
min_y = filtered['latitude'].min()
max_y = filtered['latitude'].max()

# Create a Polygon object from the bounding box
bbox_polygon = Polygon([(min_x, min_y), (min_x, max_y), (max_x, max_y), (max_x, min_y)])

# Now generate the Poisson random point pattern using the polygon
random_pattern = random.poisson(bbox_polygon, size=len(filtered))

```

### 3.6. Alpha Shape (alpha_shape): 

This represents the shape formed by the given set of points, and it can be seen as a flexible "boundary" that encompasses these points. Depending on the context of your data, this might provide insights into regions of interest, clusters, or other spatial phenomena.

    Alpha Value (alpha): 
This value gives an indication of the granularity of the computed shape. If you find this value to be small, it means the shape is capturing more detailed features of the point distribution. A larger value would mean a more generalized shape.

    Circles (circs):
These are the circles used in the Delaunay triangulation to compute the alpha shape. Each circle encompasses a subset of the points, and its size is determined by the computed alpha value. These circles can be useful for visualizing how the alpha shape is formed.

```python
# calculating alpha, alpha_shape and circs
alpha_shape, alpha, circs = libpysal.cg.alpha_shape_auto(
    filtered[['longitude', 'latitude']].values, return_circles=True
)
```

Below we visualize the borders of the whole destroyed/damaged data in our data set.

```python

# Convert alpha_shape to a GeoDataFrame
gdf_alpha_shape = gpd.GeoDataFrame(geometry=[alpha_shape], crs="EPSG:4326")

# Create a new plot
fig, ax = plt.subplots(figsize=(10, 10))

# Plot the alpha shape
gdf_alpha_shape.boundary.plot(ax=ax, color='blue', linewidth=1.5, label='Alpha Shape')

# Plot the original data points
ax.scatter(filtered['longitude'], filtered['latitude'], s=10, color='red', label='Data Points')

# If you want to plot the circles as well, you can do it like this:
for circle in circs:
    circle_x, circle_radius = circle
    circle_plot = plt.Circle((circle_x, circle_radius), circle_radius, fill=False, edgecolor='green', linestyle='--')
    ax.add_artist(circle_plot)




ax.legend()
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Alpha Shape and Data Points')
plt.grid(True)
plt.show()
```

### 3.7. G_test, F_test:
    G_test :
Observed Distribution: 
This is derived from our
filtered data points. The function calculates a test statistic based on the actual
spatial arrangement of these points. It considers how close the points are to each other within a specified distance (support parameter).

Expected Distribution under CSR: 
This is a theoretical distribution that assumes our
data points are distributed randomly in space,with no patterns, clusters, or regularities. The function often simulates numerous random distributions (point patterns) to calculate an expected test statistic under the assumption of CSR.

Comparison: 
The g_test function will then compare the observed test statistic to the expected one. If there's a significant difference between the two, it indicates that our
 data points are not randomly distributed.

The results from this test can show:

    If the points are randomly distributed (no significant difference between observed and expected).
    If the points are clustered (closer together than we'd expect by chance).
    If the points are dispersed (further apart than we'd expect by chance).

The key metric resulting from this comparison is often a p-value. A small p-value (typically less than 0.05) suggests that the observed spatial pattern is statistically significantly different from randomness.

    F_test:

The f_test function, as we've described it within the context of spatial statistics, is another method to analyze the spatial distribution of points. Just like the g_test, the f_test evaluates the spatial pattern of data points compared to a hypothetical random distribution (Complete Spatial Randomness or CSR).

Purpose: 
The function analyzes the spatial pattern of our points to see if they are randomly distributed in space or if they exhibit clustering or dispersion.

Inputs:
The array of points (filtered.values in our case).
support: 
Likely a distance threshold or maximum distance for considering neighboring points.
keep_simulations: 
Indicates that the function will store the results of simulated random patterns for later comparison.

Outputs:
Just like with g_test, the outputs would typically include:
Observed value of the test statistic based on our data.
Expected value under the assumption of CSR.
p-value, showing the probability of getting the observed pattern (or more extreme) if the points were truly distributed randomly.
Simulated values, if keep_simulations is set to True.

Interpretation:
If the observed statistic is notably different from the expected value, then the points aren't randomly distributed. They could be clustered (closer together than we'd expect under randomness) or dispersed (further apart than we'd expect).
Examining the simulated values (if kept) will allow us to see how the observed distribution compares visually to random simulations.

While both the g_test and f_test evaluate spatial point patterns, they might calculate different statistics or use different methodologies to compare the observed distribution to that of random patterns. The exact differences would depend on the library or package's specific implementation.




```python
#calculating g_test

g_test = distance_statistics.g_test(
    filtered[['longitude', 'latitude']].values, support=40, keep_simulations=True
)
```

```python
print(g_test)
```

Low p_values suggest that data points are not randomly distributed in space. They exhibit a spatial pattern that deviates significantly from what would be expected under complete spatial randomness (CSR). The exact nature of this deviation (whether it's clustering or dispersion) would depend on the specific values and trend of the G-test statistic.

```python
%matplotlib inline
f, ax = plt.subplots(
    1, 2, figsize=(9, 3), gridspec_kw=dict(width_ratios=(6, 3))
)
# plot all the simulations with very fine lines
ax[0].plot(
    g_test.support, g_test.simulations.T, color="k", alpha=0.1
)
# and show the average of simulations
ax[0].plot(
    g_test.support,
    np.median(g_test.simulations, axis=0),
    color="cyan",
    label="median simulation",
)


# and the observed pattern's G function
ax[0].plot(
    g_test.support, g_test.statistic, label="observed", color="red"
)

# clean up labels and axes
ax[0].set_xlabel("distance")
ax[0].set_ylabel("% of nearest neighbor\ndistances shorter")
ax[0].legend()
ax[0].set_xlim(0, 0.035)
ax[0].set_title(r"Ripley's $G(d)$ function")

# plot the pattern itself on the next frame
ax[1].scatter(*filtered.values.T)

# and clean up labels and axes there, too
ax[1].set_xticks([])
ax[1].set_yticks([])
ax[1].set_xticklabels([])
ax[1].set_yticklabels([])
ax[1].set_title("Pattern")
f.tight_layout()
plt.show()
```

On graph we also observe observed values are over medium simulation this also suggest clustering.

```python
#calculating f_test
f_test = distance_statistics.f_test(
    filtered.values, support=40, keep_simulations=True
)
```

```python
print(f_test)
```

Observed p values are significantly low.This means that at that particular distance ddd, the observed point pattern significantly deviates from a random pattern. It could be showing either clustering or dispersion, depending on where the F(d)lies relative to the F(d)=dF(d)  line.

```python
#graphing the Ripley's K function
f, ax = plt.subplots(
    1, 2, figsize=(9, 3), gridspec_kw=dict(width_ratios=(6, 3))
)

# plot all the simulations with very fine lines
ax[0].plot(
    f_test.support, f_test.simulations.T, color="k", alpha=0.01
)
# and show the average of simulations
ax[0].plot(
    f_test.support,
    np.median(f_test.simulations, axis=0),
    color="cyan",
    label="median simulation",
)


# and the observed pattern's F function
ax[0].plot(
    f_test.support, f_test.statistic, label="observed", color="red"
)

# clean up labels and axes
ax[0].set_xlabel("distance")
ax[0].set_ylabel("% of nearest point in pattern\ndistances shorter")
ax[0].legend()
ax[0].set_xlim(0, 0.035)
ax[0].set_title(r"Ripley's $F(d)$ function")

# plot the pattern itself on the next frame
ax[1].scatter(*filtered.values.T)

# and clean up labels and axes there, too
ax[1].set_xticks([])
ax[1].set_yticks([])
ax[1].set_xticklabels([])
ax[1].set_yticklabels([])
ax[1].set_title("Pattern")
f.tight_layout()
plt.show()
```

Ripley's F(d)F(d)F(d) function helps you identify scales at which a point pattern departs from randomness, whether it's showing clustering, dispersion, or behaving as expected under CSR.

Local peaks and troughs in the F(d)F(d)F(d) curve can indicate scales at which clustering or dispersion are particularly pronounced.
