```python
import pandas as pd
from geopy.distance import geodesic
from scipy.stats import spearmanr
import geopandas as gpd

from shapely.geometry import Polygon, Point
```

```python
def polygon_to_point(polygon):
    centroid = polygon.centroid
    latitude = centroid.y
    longitude = centroid.x
    return Point(longitude, latitude)

def point_to_coords(point):
    latitude = point.y
    longitude = point.x
    return (latitude, longitude)


geo_city['centroid'] = geo_city['geometry'].apply(polygon_to_point)
geo_city['latitude'], geo_city['longitude'] = zip(*geo_city['centroid'].map(point_to_coords))


geo_city['Location_name'] = geo_city['Province'] + '-' + geo_city['city'] 
```

```python
def clean_geodata(gdf):
    # Check for missing values
    gdf = gdf.dropna()

    # Check for duplicates
    gdf = gdf.drop_duplicates()

    # Check for invalid geometries
    gdf = gdf[gdf.geometry.is_valid]

    # Check for inconsistent CRS
    if not gdf.crs:
        gdf.crs = {'init': 'epsg:4326'}  # Set CRS if not set already
        gdf = gdf.to_crs({'init': 'epsg:4326'})  # Reproject to common CRS
    return gdf
clean_geodata(geo_city)
```
